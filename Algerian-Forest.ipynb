## Data Cleaning

```python
## Missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index. can make a new column ased on the Region

2: "Sidi-bel Abbes Region DAtaset"
Add new column with region

```python
## Adding new Column with Region
dataset.loc[:122,"Region"] = 0
dataset.loc[122:, "Region"] =1
```

```python
dataset.info()
```

```python
dataset['Region'] = dataset['Region'].astype(int)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
## Removing the null Value
dataset = dataset.dropna().reset_index(drop=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.iloc[[122]]
```

```python
## Unnesseary Row 
dataset= dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns
```

```python
## removing white Spaces in column names
dataset.columns= dataset.columns.str.strip()
```

```python
dataset.columns
```

### Changes the required column as Integer Data Type

```python
dataset[['day', 'month', 'year', 'Temperature', 'RH']] = dataset[['day', 'month', 'year', 'Temperature', 'RH']].astype(int)
```

```python
dataset.info()
```

## Changing the other Columns to float and Integer data type

```python
object = [features for features in dataset.columns if dataset[features].dtypes =='O']
```

```python
for i in object:
    if i != 'Classes':
        dataset[i] = dataset[i].astype(float)
```

```python
dataset.info()
```

### Saving the clean dataset

```python
dataset.to_csv('Algerian_forest_fires_dataset_Cleaned.csv', index=False)
```
